```python
# Load packages
import os.path as op
import glob
import mne
import numpy as np
import pandas as pd
```

```python
# Define data paths
root_dir = 'C:/Users/mvmigem/Documents/data/project_2/'
epo_data_dir = op.join(root_dir,"preprocessed/mastoid-epo/unpaired/*")
behav_data_dir = op.join(root_dir,"aligned_behav/*")
destination_dir = op.join(root_dir,"preprocessed/mastoid-epo/")
# Get individual paths
epo_paths = glob.glob(epo_data_dir)
behav_paths = glob.glob(behav_data_dir)
# Adjust for dropout
select_bpaths = [behav_paths[x] for x in range(len(behav_paths)) if x not in [26,30,32]]
```

```python
# Remove dropped events from the metadata and add metadata to the epoch
for epath,bpath in zip(epo_paths,select_bpaths):
    sub = int(epath.split('-')[3])
    epochs = mne.read_epochs(epath)
    events = epochs.events
    meta_data = pd.read_csv(bpath)
    drp_log = list(epochs.drop_log)
    drp_log = [x for x in drp_log if x !=('IGNORED',)]
    dropped_epochs = [i for i, val in enumerate(drp_log) if val != ()]
    meta_data = meta_data.drop(meta_data.index[dropped_epochs])
    # Check if correct
    if all(events[:,2] == meta_data['eeg_trigger'].to_numpy()):
        epochs.metadata = meta_data
        epochs.save(destination_dir + f"paired-mastoid-sub-{sub:02}-epo.fif", overwrite = True)
    else:
        print(f"sub {sub} not aligned")
```
