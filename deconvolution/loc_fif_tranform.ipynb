```python
import mne, os
import pandas as pd


input_path = 'C:/Users/mvmigem/Documents/data/project_2/preprocessed/localiser/mastoid-raw/'
output_path = 'C:/Users/mvmigem/Documents/data/project_2/preprocessed/localiser/mastoid-raw-csv/'
# Load the 10-20 montage
montage = mne.channels.make_standard_montage('biosemi64')
# Get the list of channel names
electrode_names = montage.ch_names[:]

roi_channels = ['Oz','O1','O2',
                'PO3','POz','PO4',
                'P1','Pz','P2',
                'CP1','CPz','CP2',
                ]
```

```python
for i in roi_channels:
    spec_path  = os.path.join(output_path, f'raw-{i}')
    if os.path.isdir(spec_path):
        continue

    el_dir  = os.makedirs(spec_path)
    el = i
    for sub in range(1,36):
        if sub in [27,31]: # has no data
            continue
        # Load and read files
        clean_raw_path = os.path.join(input_path,f'sub-{sub:02}-localiser-raw.fif')
        raw = mne.io.read_raw_fif(clean_raw_path)
        raw.info['bads'] = []
        # Copy event info
        events = mne.find_events(raw,shortest_event=1)
        # Event dict
        event_id = {        # This needs to be short in the online preprocess only 4 trigger markers (4 quads)
                    'position1':80,'position2':81, 'position3':82,'position4':83, 
        }
        # Select the electrode
        raw_el_df =  raw.copy().pick([el]).to_data_frame(long_format = False)
        # selected_el = peak_properties[peak_properties['subject']==subject]['all_pos_peak_channel'].iloc[0]
        # raw_select_df = raw.copy().pick([selected_el]).to_data_frame(long_format = False)
        # Assemble the event df with extra info
        events_df = pd.DataFrame(events[:,0],columns=['sample'])
        events_df['event_codes'] = events[:,2]
        events_df['position'] = (events_df['event_codes'] %  10) +1
        events_df['subject'] = sub

        # Save raws
        raw_el_df_path = os.path.join(output_path,f'raw-{el}/raw-mastoid-{el}-{sub:02}.csv')
        # raw_select_df_path = os.path.join(destinantion_path,f'raw-selected/raw_mastoidref_selected_{sub:02}.csv')
        raw_el_df.to_csv(raw_el_df_path,index = False)
        # raw_select_df.to_csv(raw_select_df_path,index = False)
        # Save events 
        events_csv_path = os.path.join(output_path,f'events/events-sub-{sub:02}.csv')
        events_df.to_csv(events_csv_path,index = False)
```
